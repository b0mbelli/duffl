```python
import pandas as pd
import numpy as np
from collections import Counter

df= pd.read_csv(r"C:\Users\Luca\OneDrive\Desktop\Duffl\Whales Data\Whale_Zeros_School_2024.csv")
df.head()
```

```python
#filtering for delivery time
delivery_over_30 = df['delivery_time'] > 30
df[delivery_over_30]
```

```python
#filtering for refunds
item_refunded = df['amount_refunded'] > 0
df[item_refunded]
```

```python
#for both refund and late delivery
df[item_refunded & delivery_over_30]
```

```python
#adding in rainy day filter
rainy_day = df['Rainy_day'] == 1
df[rainy_day]
```

```python
#packer error counts
packer_counts = df.value_counts(subset = "Packer")
print(packer_counts)
```

```python
#racer error counts
racer_counts = df.value_counts(subset = "racer")
print(racer_counts)
```

```python
#general data
print("number orders with a refund is", df[df['amount_refunded'] > 0]['amount_refunded'].count())
print("number orders with delivery time over 30 mins is", df[df['delivery_time'] > 30]['delivery_time'].count())
print("number orders with delivery time over 30 mins is", df[(df['delivery_time'] > 30) & (df['Rainy_day'] != 1)]['delivery_time'].count())
print("number orders with both a refund and a delivery time over 30 mins is", df[(df['delivery_time'] > 30) & (df["amount_refunded"] > 0)]['amount_refunded'].count())
print("number orders with the above factors that was not a rainy day is", df[(df['delivery_time'] > 30) & (df["amount_refunded"] > 0) & (df['Rainy_day'] != 1)]['delivery_time'].count())
```

```python
#the most impacted customers
most_boned = df.value_counts(subset = "customer_name")
print('the whales getting the most boned are:\n', most_boned)
most_boned_id = df.value_counts(subset = 'customer_id')
print('\nby id:\n', most_boned_id)
```
